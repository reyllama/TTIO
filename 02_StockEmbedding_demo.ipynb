```
import pandas as pd
import numpy as np
import torch
from torch import nn, optim
from torch.autograd import Variable
from google.colab import files
uploaded = files.upload()
```

```
import io
df = pd.read_csv(io.BytesIO(uploaded['cor1.csv']))
df.head()
```

```
dtype = torch.FloatTensor
epochs = 1
emb_dim = 32
embeddings = Variable(torch.randn(emb_dim, len(df)).type(dtype), requires_grad=True)
lr = 0.003
optimizer = optim.Adam([embeddings], lr=lr)

for epoch in range(epochs):
  loss = 0
  for i in range(len(df)):
    for j in range(i+1, len(df)):
      rho = df.iloc[i,j]
      optimizer.zero_grad()
      loss += np.abs(rho)*torch.norm(embeddings[:, i]-embeddings[:, j]) - (1-np.abs(rho))*torch.norm(embeddings[:, i]-embeddings[:, j])
  loss.backward(retain_graph=True)
  optimizer.step()
      

embeddings[:, 3:6]
```
